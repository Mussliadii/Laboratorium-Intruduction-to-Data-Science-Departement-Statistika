```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.stats as stats
import statsmodels.stats.api as sms
import statsmodels.formula.api as smf
from sklearn.linear_model import LinearRegression
```

```python
datamus = pd.read_excel('D:/All Data I Have/Data lab 5 psd.xlsx')
datamus #MUS
```

```python
x = datamus ['MYS']
y = datamus ['IPM']
```

```python
x = sm.add_constant (x)
result_mus = sm.OLS(y,x).fit()
print(result_mus.summary())
```

```python
model=sm.OLS(y,x).fit()
predictions=model.predict(x)
model.summary()
```

```python
def lzip(a, b):
    return list(zip(a, b))
```

```python
#homoskedastitas
fit =smf.ols('IPM ~ MYS', data=datamus).fit()
names =['Largrange multiplier statistic', 'p-value', 'f-value', 'f-value']
test_result = sms.het_breuschpagan(fit.resid, fit.model.exog)
lzip(names,test_result)
```
