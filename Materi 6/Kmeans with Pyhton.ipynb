```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
#MUS
datamus = pd.read_clipboard()
datamus.head(5)
```

```python
scaler = StandardScaler()
scaler.fit(datamus)
scaled = scaler.transform(datamus)
datamus_scaled = pd.DataFrame(scaled, columns=[ 'UAS', 'UTS', 'Tugas 1'])

datamus_scaled.head()
```

```python
datamus = datamus[['UTS', 'UAS', 'Tugas 1']]
wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init='k-means++')
    kmeans.fit(datamus)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))
plt.plot(range(1,11), wcss, linewidth=2, marker= '8')
plt.xticks(np.arange(1,11,1))
plt.xlabel('K Values')
plt.ylabel('WCSS')
plt.show()
```

```python
#MUS
mw=KMeans(n_clusters=3)
mw.fit(datamus)
y=mw.predict(datamus)
datamus['label'] = y
datamus.head(10)
```
