```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('datalaporan 1.csv',sep=";")
df
```

```python
df['Glucose'].describe()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(data=df, y='Glucose', color='skyblue')
plt.title("Boxplot Glucose", fontsize=15)
plt.xlabel("Glucose", fontsize=12)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Menggunakan Seaborn untuk mengatur gaya plot
sns.set(style="whitegrid")

# Membuat histogram untuk variabel "BloodPressure"
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Age', bins=10, color="cadetblue", edgecolor="black")

# Menambahkan judul dan label sumbu
plt.title("Distribusi Age", fontsize=15)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Frekuensi", fontsize=12)

# Menampilkan histogram
plt.show()
```

```python
# Menampilkan count plot
#Alul
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(data=df, x='Outcome')
plt.title('Distribusi Kasus')
plt.xlabel('Kasus')
plt.ylabel('Jumlah')
plt.show()

# Menampilkan nilai-nilai unik dan jumlahnya
print(df['Outcome'].value_counts())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Contoh menggunakan data_giya1
sns.scatterplot(x=df['Glucose'], y=df['Age'], color='cadetblue')

# Menambahkan judul dan label sumbu
plt.title("Scatterplot Glucose vs Age", fontsize=15)
plt.xlabel("Glucose", fontsize=12)
plt.ylabel("Age", fontsize=12)

# Menampilkan scatterplot
plt.show()
```

```python
#Persentase nasabah existing and attrited
#Alul
att_count = df.Outcome.value_counts().reset_index()
att_count.columns=["Outcome", "Counts"]

fig, ax = plt.subplots()
ax.pie(att_count.Counts, labels = att_count.Outcome, autopct ='% .1f%%', startangle = 30)
ax.set_title('Persentase Kasus')
```
